```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('September Feedback.csv')
```

```python
location_data = df['Location'].value_counts()
```

```python
location_data.plot(kind='barh',color='orange',xlabel='Frequency',title='Fintech Locations')
```

```python
location_data.plot(kind='pie',autopct='%1.1f%%',legend=False,title='September Location Results')
```

Singapore is a most popular location within the fintech company and the second biggest location is Cardiff. 

```python
team_data = df['Team'].value_counts()
```

```python
team_data.plot(kind='barh',color='blue',xlabel='Frequency',title='Fintech Teams')
```

```python
team_data.plot(kind='pie',autopct='%1.1f%%',legend=False,title='September Team Results')
```

```python
department_data = df['Department'].value_counts()
```

```python
department_data.plot(kind='barh',color='green',xlabel='Frequency',title='Fintech Department')
```

Technology is a most popular department within the fintech industry.  Growth is a smallest departments within the fintech industy. 

```python
department_data.plot(kind='pie',autopct='%1.1f%%',title='September Department Results',legend=False)
```
