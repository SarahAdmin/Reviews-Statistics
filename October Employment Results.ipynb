```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('October Feedback.csv')
```

```python
location_data = df['Location'].value_counts()
```

```python
location_data.plot(kind='barh',color='yellow',xlabel='Frequency',title='Fintech Locations')
```

```python
location_data.plot(kind='pie',legend=False,autopct='%1.1f%%',title='October Location Results')
```

Singapore is a most popular location within the fintech industry. The second biggest location is London (Remote) and the third biggest location is UK (Remote). Other locations has the same frequencies.

```python
team_data = df['Team'].value_counts()
```

```python
team_data.plot(kind='barh',color='coral',xlabel='Frequency',title='Fintech Teams')
```

```python
team_data.plot(kind='pie',autopct='%1.1f%%',legend=False,title='October Team Results')
```

```python
department_data = df['Department'].value_counts()
```

```python
department_data.plot(kind='barh',color='crimson',xlabel='Frequency',title='Fintech Teams')
```

Sales is the most popular department within the fintech company. The second biggest department is Banking. Other departments has same frequencies. 

```python
department_data.plot(kind='pie',autopct='%1.1f%%',legend=False,title='October Department Results')
```
