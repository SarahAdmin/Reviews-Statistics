```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('December Feeback.csv')
```

```python
result1 = df['Location'].value_counts()
```

```python
result1.plot(kind='barh',title='December Locations',xlabel='Frequency',color='red')
```

The most popular location within the company is 'USA (Remote)' because it has the largest number of teams and departments. 

```python
result2 = df['Department'].value_counts()
```

```python
result2.plot(kind='barh',title='December Departments',xlabel='Frequency',color='orange')
```

The most popular department is Sales because it is a biggest number of roles. However, other departments has been decreasing the number of roles.

```python
result3 = df['Team'].value_counts()
```

```python
result3.plot(kind='barh',title='December Teams',xlabel='Frequency',color='indigo')
```

The most popular team is 'Customer Team' because it has biggest number of roles.
