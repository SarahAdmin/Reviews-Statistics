```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Jan Feedback.csv')
```

```python
location_data = df['Location'].value_counts()
```

```python
location_data.plot(kind='pie',autopct='%1.1f%%',title='January Location Results')
```

```python
location_data.plot(kind='barh',color='orange',title='Fintech Locations',xlabel='Frequency')
```

The biggest location is United Kingdom because it has most popular teams and departments. However, other locations has the same frequencies. 

```python
team_data = df['Team'].value_counts()
```

```python
team_data.plot(kind='pie',autopct='%1.1f%%',title='January Team Results',legend=False)
```

```python
team_data.plot(kind='barh',color='deeppink',title='Fintech Teams',xlabel='Frequency')
```

These teams in January has the same frequencies while increasing the number of teams.

```python
department_data = df['Department'].value_counts()
```

```python
department_data.plot(kind='pie',autopct='%1.1f%%',title='January Department Results',legend=False)
```

```python
department_data.plot(kind='barh',color='navy',xlabel='Frequency',title='Fintech Departments')
```

The biggest department was Sales within the fintech company. The second biggest department was Compliance. The smallest department is Banking.
