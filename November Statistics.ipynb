```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('November Feedback.csv')
```

```python
result1 = df['Location'].value_counts()
```

```python
result1.plot(kind='barh',color='rebeccapurple',title='Fintech Locations',legend=False)
```

The most popular location is shown on the bar graph is 'Singapore (Remote)' because the company has the highest teams at Singapore. Also, the second location is 'UK (Remote)'. Least popular is United Kingdom (Hybrid) with the smallest number.

```python
result2 = df['Department'].value_counts()
```

```python
result2.plot(kind='barh',color='red',title='November Departments')
```

The most popular department within the fintech company is Sales has a biggest number of positions. The second department is Operations with the number of positions. The least popular is Banking with the smallest number of positions.

```python
result3 = df['Team'].value_counts()
```

```python
result3.plot(kind='barh',color='violet',title='November Teams')
```

The most popular team within the fintech company is Product because it has a largest number of roles. However, other roles has been increasing the number of roles.
